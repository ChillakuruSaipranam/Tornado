```python
import pandas as pd 
import matplotlib.pyplot as plt 

df = pd.read_csv("1950-2023_all_tornadoes.csv")
df.head()
```

```python
list(df.columns)
```

```python
max_len = df.groupby('yr')['len'].max().reset_index()

print(max_len)
max_len.to_csv('max_len.csv',index = False)

plt.figure(figsize=(8, 6))

plt.plot(max_len['yr'], max_len['len'], color='blue')

plt.title('Line plot of Year vs Maximum length')
plt.xlabel('yr')
plt.ylabel('max_len')
plt.show()
```
